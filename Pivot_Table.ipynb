# IMPORTING FILE 

```python
df=pd.read_csv('/content/Practicum_Datasets/world_cities_air_water_quality.csv')
```

```python
df
```

# CHANGING COLUMN NAMES

```python
columns = ['City','Region','Country','AirQuality','WaterPollution']
df.columns=columns
```

```python
df.head(2)
```

# CLEANING COLUMN ELEMENTS FROM ' " '

```python
df['Region']=df['Region'].map(lambda x: x.replace('"',"") )
```

```python
df.head(2)
```

```python
df['Country']=df['Country'].map(lambda x: x.replace('"',"") )
```

```python
df.head()
```

# REMOVING SPACES FROM ELEMENTS

```python
df['Country']=df['Country'].map(lambda x: x.strip())
```

```python
df['Country'][3]
```

# PIVOT_TABLE 

```python
df=pd.pivot_table(df, index=['Country','City'])
```

```python
df
```

```python
df.to_csv('Water_pollution.csv')
```

```python
df
```

```python
df['AirPollution']=100-df['AirQuality']
```

```python
df
```

```python
df['WaterQuality']=100-df['WaterPollution']
```

```python
df
```

```python
df1=df.copy()
```

```python
df3=pd.DataFrame(df1.values, index=df1.index, columns=[['WATER','WATER','AIR','AIR'],['WaterPollution','WaterQuality','AirPollution','AirQuality']])
```
